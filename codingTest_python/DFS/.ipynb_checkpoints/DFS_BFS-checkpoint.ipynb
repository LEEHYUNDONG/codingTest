```python
# 경쟁적 전염 품
from collections import deque

dx = [0, 1, 0, -1]
dy = [1, 0, -1, 0]
n, k = map(int, input().split())
graph = []
for i in range(n):
    graph.append(list(map(int, input().split())))

s, x, y = map(int, input().split())

def bfs(rx, ry, s):
    cnt = 0
    time = 0
    max_time = 0
    tmp = []
    for i in range(n):
        for j in range(n):
            if graph[i][j] != 0:
                tmp.append([graph[i][j], i, j, time])
                    
    tmp.sort()
    queue = deque(tmp)
    
    while queue:
        v, x, y, time = queue.popleft()
        if time == s:
            break
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 <= nx < n and 0 <= ny < n:
                if graph[nx][ny] == 0:
                    graph[nx][ny] = graph[x][y]
                    queue.append([v, nx, ny, time+1])
                    cnt += 1  

    return graph[rx][ry]

print(bfs(x-1, y-1, s))
```

```python
# 연산자 끼워넣기
from collections import deque
from itertools import permutations

n = int(input())
num = list(map(int, input().split()))
op = list(map(int, input().split()))
oper = []

for i in range(len(op)):
    for j in range(op[i]):
        if i == 0:
            oper.append('+')
        elif i == 1:
            oper.append('-')
        elif i == 2:
            oper.append('*')
        elif i== 3:
            oper.append('/')

queue = deque(list(permutations(oper, n-1)))

maxNum = -1000000000
minNum = 1000000000
tot = 0

while queue:
    op = list(queue.popleft())
    tot = num[0]
    for i in range(1, len(num)):
        operation = op.pop()
        if operation == '+':
            tot += num[i]
        elif operation == '-':
            tot -= num[i]
        elif operation == '*':
            tot *= num[i]
        elif operation == '/':
            tot = int(tot/num[i])
    if tot > maxNum:
        maxNum = tot
    if tot < minNum:
        minNum = tot
print(maxNum, minNum)
```

```python
# 감시 피하기
# 품
from collections import deque
from itertools import combinations
import copy

# test_case = [
#     ['X', 'S', 'X', 'X', 'T'],
#     ['T', 'X', 'S', 'X', 'X'],
#     ['X', 'X', 'X', 'X', 'X'],
#     ['X', 'T', 'X', 'X', 'X'],
#     ['X', 'X', 'T', 'X', 'X'],
    
# ]
n = int(input())
test_case = [list(input().split()) for _ in range(n)]

vacant = []
student = 0

for i in range(n):
    for j in range(n):
        if test_case[i][j] == 'X':
            vacant.append([i, j])
        if test_case[i][j] == 'S':
            student += 1

def bfs(t, student):
    s = student
    graph = copy.deepcopy(t)
    q = deque([])
    for i in range(n):
        for j in range(n):
            if graph[i][j] == 'T':
                q.append([i, j])
    while q:
        x, y = q.popleft()
        nx = x
        ny = y
        if 0 <= nx < n and 0 <= ny < n:
            for j in range(nx-1, -1, -1):
                if graph[j][ny] == 'O':
                    break
                graph[j][ny] = 'T'
            for j in range(nx, n):
                if graph[j][ny] == 'O':
                    break
                graph[j][ny] = 'T'
            for j in range(ny-1, -1, -1):
                if graph[nx][j] == 'O':
                    break
                graph[nx][j] = 'T'
            for j in range(ny, n):
                if graph[nx][j] == 'O':
                    break
                graph[nx][j] = 'T' 
    
    cnt = 0
    for i in range(n):
        for j in range(n):
            if graph[i][j] == 'S':
                cnt += 1
    
    if cnt == student:
        return True
    else:
        return False

flag = False

for data in combinations(vacant, 3):
    for x, y in data:
        test_case[x][y] = 'O'
    if bfs(test_case, student):
        flag = True       
        break
    for x, y in data:
        test_case[x][y] = 'X'

if flag:
    print('YES')
else:
    print('NO')
```

```python
# 인구 이동
from collections import deque

# 땅의 크기(N), L, R 값을 입력받기
n, l, r = map(int, input().split())

# 전체 나라의 정보(N x N)를 입력 받기
graph = []
for _ in range(n):
    graph.append(list(map(int, input().split())))

dx = [-1, 0, 1, 0]
dy = [0, -1, 0, 1]

# 특정 위치에서 출발하여 모든 연합을 체크한 뒤에 데이터 갱신
def process(x, y, index):
    # (x, y)의 위치와 연결된 나라(연합) 정보를 담는 리스트
    united = []
    united.append((x, y))
    # 너비 우선 탐색 (BFS)을 위한 큐 라이브러리 사용
    q = deque()
    q.append((x, y))
    union[x][y] = index # 현재 연합의 번호 할당
    summary = graph[x][y] # 현재 연합의 전체 인구 수
    count = 1 # 현재 연합의 국가 수
    # 큐가 빌 때까지 반복(BFS)
    while q:
        x, y = q.popleft()
        # 현재 위치에서 4가지 방향을 확인하며
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            # 바로 옆에 있는 나라를 확인하여
            if 0 <= nx < n and 0 <= ny < n and union[nx][ny] == -1:
                # 옆에 있는 나라와 인구 차이가 L명 이상, R명 이하라면
                if l <= abs(graph[nx][ny] - graph[x][y]) <= r:
                    q.append((nx, ny))
                    # 연합에 추가하기
                    union[nx][ny] = index
                    summary += graph[nx][ny]
                    count += 1
                    united.append((nx, ny))
    # 연합 국가끼리 인구를 분배
    for i, j in united:
        graph[i][j] = summary // count

total_count = 0

# 더 이상 인구 이동을 할 수 없을 때까지 반복
while True:
    union = [[-1] * n for _ in range(n)]
    index = 0
    for i in range(n):
        for j in range(n):
            if union[i][j] == -1: # 해당 나라가 아직 처리되지 않았다면
                process(i, j, index)
                index += 1
    # 모든 인구 이동이 끝난 경우
    if index == n * n:
        break
    total_count += 1

# 인구 이동 횟수 출력
print(total_count)
```
